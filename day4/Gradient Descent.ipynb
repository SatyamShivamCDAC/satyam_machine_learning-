```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import log_loss,classification_report
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from datetime import datetime
```

<h1 style = 'color:orange'>Gradient Descent</h1>

```python
#one iteration
hr = pd.read_csv('../Cases/HRAnalytics/HR_comma_sep.csv')

w = 0.1 #weight 
b = 0.1 #bias

x = np.array(hr['satisfaction_level']) #independent variable
y = np.array(hr['left']) #dependent variable
n = len(y)
alpha = 0.1  #alpha
y_cap = 1/(1 + np.exp((w*x + b)*-1)) ##yi_cap/ sigmoid funtiom

L = log_loss(y,y_cap) # loss

wd = ((y-y_cap)*x).mean() #derivate of w
bd = ((y-y_cap)).mean()

# first iteration/ epochs
w = w - alpha*wd
b = b - alpha*bd

w,b
```

<h1 style = 'color:Orange'>Kyphosis Dataset</h1>

```python
kyph = pd.read_csv('../Cases/Kyphosis/Kyphosis.csv')

le = LabelEncoder()
sampler = MinMaxScaler().set_output(transform='pandas')

kyph['Kyphosis'] = le.fit_transform(kyph['Kyphosis'])

y = kyph['Kyphosis']
X = kyph.drop('Kyphosis' , axis = 1)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=25,stratify=y)

X_train_sampled = sampler.fit_transform(X_train)
X_test_sampled = sampler.transform(X_test)

x1,x2,x3 = X_train_sampled['Age'],X_train_sampled['Number'],X_train_sampled['Start']

w1,w2,w3,b,eta = 0.1,0.1,0.1,0.1, 0.3

for i in range(10000):
    y_cap = 1/(1+np.exp((w1*x1 + w2*x2 + w3*x3 + b)*-1))

    L = log_loss(y_train,y_cap)
    bd = -np.mean((y-y_cap))
    wd1 = -np.mean((y-y_cap)*x1)
    wd2 = -np.mean((y-y_cap)*x2)
    wd3 = -np.mean((y-y_cap)*x3)

    w1 = w1 - eta*wd1
    w2 = w2 - eta*wd2
    w3 = w3 - eta*wd3
    b = b - eta*bd

    print(f'w1 = {w1:.2},w2 = {w2:.2},w3 = {w3:.2},loss = {L:.2}')
    
```

```python
#checkin for test set
y_pred_proba = 1/(1+np.exp((w1*X_test_sampled['Age'] + w2*X_test_sampled['Number'] + w3*X_test_sampled['Start'] + b)*-1))
print(f'Loss for test set {log_loss(y_test,y_pred_proba)}')
y_pred = np.where(y_pred_proba>0.5,1,0)
print(classification_report(y_test,y_pred))
```

<h1 style = 'color:orange'>Batch Gradient Descent</h1>

```python
x = np.array([1.2,1.4,0.5,0.6,1.1,1.3])
y = np.array([1,1,0,0,1,1])
time = datetime.now()
w,b = 0.1,0.1
eta = 0.5
tol = 0.01 # tolerance

for i in range(100000):
    y_cap = 1/(1+np.exp((w*x + b)*-1))
    L = log_loss(y,y_cap)
    
    if(L<tol):
        print(f'Iterations - {i}')
        print(f'Loss - {L}')
        break
    wd = -np.mean((y-y_cap)*x)
    bd = -np.mean((y-y_cap))
    w = w - eta*wd
    b = b - eta*bd
print(f'Time take = {datetime.now() - time}')
```

<h1 style= 'color:Orange'>Stochastic Gradient Descent</h1>

```python
x = np.array([1.2,1.4,0.5,0.6,1.1,1.3])
y = np.array([1,1,0,0,1,1])
time = datetime.now()
w,b = 0.1,0.1
eta = 0.5
tol = 0.01 # tolerance

for i in range(100000):
    for j in range(len(x)):
        y_cap =  1/(1+np.exp((w*x[j] + b)*-1))
        db = -(y[j] - y_cap)
        dw = -((y[j] - y_cap)*x[j])
        w = w - eta*dw
        b = b - eta*db
    
    y_cap_aLL =  1/(1+np.exp((w*x + b)*-1))
    L = log_loss(y, y_cap_aLL)
    if(L<tol):
        print(f'Iterations - {i*j}')
        print(f'Loss - {L}')
        break
print(f'Time take = {datetime.now() - time}')
```

<h1 style= 'color:Orange'>Min-Batch Gradient Descent</h1>

```python
x = np.array([1.2,1.4,0.5,0.6,1.1,1.3])
y = np.array([1,1,0,0,1,1])
time = datetime.now()
w,b = 0.1,0.1
eta = 0.5
tol = 0.01 # tolerance

# batches = [[0, 1, 2], [3, 4, 5]]
batches = [[0, 1], [2, 3], [4, 5]]

for i in range(100000):
    for batch in batches:
        y_cap = 1/(1+np.exp((w*x[batch] + b)*-1))
        wd = -np.mean((y[batch]-y_cap)*x[batch])
        bd = -np.mean((y[batch]-y_cap))
        w = w - eta*wd
        b = b - eta*bd
    i = i + j
    y_cap_all = 1/(1+np.exp((w*x + b)*-1))
    L = log_loss(y,y_cap_all)
    
    if(L<tol):
        print(f'Iterations - {i}')
        print(f'Loss - {L}')
        break

print(f'Time take = {datetime.now() - time}')
```
