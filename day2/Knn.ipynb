```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

<h1 style = 'color: orange'>KNN</h1>

```python
cancer = pd.read_csv('../Cases/Wisconsin/BreastCancer.csv')
y = cancer['Class']
X = cancer.drop(['Class','Code'], axis = 1)

X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,test_size=0.3,random_state=25)

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
#for different number of neighbours 
score = []
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)

    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    score.append([i,accuracy_score(y_pred,y_test)])
score = pd.DataFrame(score,columns=['k neighbours','accuracy'])
score.sort_values('accuracy',ascending=False)
```

<h1 style= 'color:Orange'>Human Resource Dataset</h1>

```python
hr = pd.read_csv('../Cases/HRAnalytics/HR_comma_sep.csv')

y = hr['left']
X = hr.drop('left',axis = 1)
X = pd.get_dummies(X,drop_first=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=25,stratify=y)
score = []
for k in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    score.append([k,accuracy_score(y_pred,y_test)])
score = pd.DataFrame(score,columns=['k neighbours','accuracy'])
score.sort_values('accuracy',ascending=False)
```

```python
#scaling hr dataset before training
scalar = StandardScaler()
X_train_scl = scalar.fit_transform(X_train)
X_test_scl = scalar.transform(X_test)

score = []

for k in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scl,y_train)
    y_pred = knn.predict(X_test_scl)
    score.append([k,accuracy_score(y_pred,y_test)])
score = pd.DataFrame(score,columns=['k','accuracy'])
score.sort_values('accuracy',ascending=False)
```
