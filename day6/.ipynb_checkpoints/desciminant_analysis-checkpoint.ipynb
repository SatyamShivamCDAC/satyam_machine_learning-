```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis,LinearDiscriminantAnalysis
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import make_column_selector,make_column_transformer 
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
```

<h1 style = 'color:Orange'>Descriminant Analysis</h1>

```python
glass = pd.read_csv('../Cases/Glass_Identification/Glass.csv')

y = glass['Type']
X = glass.drop('Type',axis=1)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=25,stratify=y)

lda = LinearDiscriminantAnalysis()
lda.fit(X_train,y_train)

y_pred = lda.predict(X_test)

print(classification_report(y_test,y_pred))
```

<h1 style = 'color:Orange'>SVM on LDA Transformation</h1>

```python
lda = LinearDiscriminantAnalysis().set_output(transform='pandas')

iris = pd.read_csv('../Datasets/iris.csv')

y = iris['Species']
X = iris.drop('Species',axis=1)

X_lda = lda.fit_transform(X,y)  #dimension reduction
X_lda['Species'] = iris['Species']

svm  = SVC(kernel='linear')

y = X_lda['Species']
X = X_lda.drop('Species',axis=1)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=25,stratify=y)
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print(classification_report(y_test,y_pred))
```
