```python
from scipy.cluster.hierarchy import linkage,dendrogram

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_scorel

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

<h1 style = 'color: orange'>Hierarcal CLustering</h1>

```python
df = pd.read_csv('../Datasets/simp_clust.csv',index_col=0)
df
plt.scatter(df['X1'],df['X2'])
for i in range(df.shape[0]):
    plt.text(x = df['X1'].values[i], y =df['X2'].values[i],s = list(df.index)[i])
plt.title("Dataset")
plt.xlabel('X1')
plt.xlabel('X2')

plt.show()
```

```python
#Average Linkage
scalar = StandardScaler().set_output(transform='pandas')
df_scaled = scalar.fit_transform(df)
link = 'average'
merging = linkage(df_scaled,method = link)  #defining merging way
dendrogram(merging,labels = list(df_scaled.index)) #plots dendogram
plt.title(link + " linkage") 

plt.xlabel('X1')
plt.xlabel('X2')
plt.show()
```

```python
#Single Linkage
scalar = StandardScaler().set_output(transform='pandas')
df_scaled = scalar.fit_transform(df)
link = 'single'
merging = linkage(df_scaled,method = link)
dendrogram(merging,labels = list(df_scaled.index))
plt.title(link + " linkage")

plt.xlabel('X1')
plt.xlabel('X2')
plt.show()
```

```python
#Complete Linkage
scalar = StandardScaler().set_output(transform='pandas')
df_scaled = scalar.fit_transform(df)
link = 'complete'
merging = linkage(df_scaled,method = link)
dendrogram(merging,labels = list(df_scaled.index))
plt.title(link + " linkage")

plt.xlabel('X1')
plt.xlabel('X2')
plt.show()
```

```python
#Centroid Linkage
scalar = StandardScaler().set_output(transform='pandas')
df_scaled = scalar.fit_transform(df)
link = 'centroid'
merging = linkage(df_scaled,method = link)
dendrogram(merging,labels = list(df_scaled.index))
plt.title(link + " linkage")

plt.xlabel('X1')
plt.xlabel('X2')
plt.show()
```

```python
#Weighted Linkage
scalar = StandardScaler().set_output(transform='pandas')
df_scaled = scalar.fit_transform(df)
link = 'weighted'
merging = linkage(df_scaled,method = link)
dendrogram(merging,labels = list(df_scaled.index))
plt.title(link + " linkage")

plt.xlabel('X1')
plt.xlabel('X2')
plt.show()
```
