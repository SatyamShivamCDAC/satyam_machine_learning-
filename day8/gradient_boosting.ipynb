```python
sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')
sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')

X = sonar.drop('Class',axis = 1)
y = sonar['Class']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=25,test_size=0.3,stratify=y)

gbc = GradientBoostingClassifier(learning_rate=0.8,n_estimators=50,max_depth=3)
gbc.fit(X_train,y_train)
y_pred = gbc.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
# with different learning rate,deapth,estimators

sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')
sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')

X = sonar.drop('Class',axis = 1)
y = sonar['Class']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=25,test_size=0.3,stratify=y)

learning_rates = np.linspace(0.01,0.8,20)
n_est = [50,100,200]
depths = [3,5,None]
scores = []

for n in n_est:
    for l in learning_rates:
        for d in depths:
            gbc = GradientBoostingClassifier(random_state=25,learning_rate=l,n_estimators=n,max_depth=d)
            gbc.fit(X_train,y_train)
            y_pred = gbc.predict_proba(X_test)
            scores.append([l,n,d,log_loss(y_test,y_pred)])
scores = pd.DataFrame(scores,columns=['Learning rate','N estimators','Deapth','log loss'])
scores.sort_values('log loss')
```
