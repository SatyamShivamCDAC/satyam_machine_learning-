```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier,AdaBoostRegressor
from sklearn.metrics import classification_report,log_loss,r2_score,mean_squared_error
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression,LinearRegression,ElasticNet
from sklearn.naive_bayes import GaussianNB
```

```python

#classification models
dtc = DecisionTreeClassifier(random_state=25,max_depth=1)
knn = KNeighborsClassifier()
nb = GaussianNB()
lR = LogisticRegression()

#Regression models
lr = LinearRegression()
dtr = DecisionTreeRegressor(random_state=25,max_depth=1)
en = ElasticNet()
```

<h1 style = color:Orange>Adaboost</h1>

```python
sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')
sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')

X = sonar.drop('Class',axis = 1)
y = sonar['Class']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=25,test_size=0.3,stratify=y)

ada = AdaBoostClassifier(estimator=nb, n_estimators=10,random_state=25)
ada.fit(X_train,y_train)

y_pred = ada.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#for different estimators

sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')
sonar = pd.read_csv('../Cases/Sonar/Sonar.csv')

X = sonar.drop('Class',axis = 1)
y = sonar['Class']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=25,test_size=0.3,stratify=y)

weak_lnr = [10,15,25]
est_list = [nb,dtc,lR]
score = []

for e in est_list:
    for o in weak_lnr:
        ada = AdaBoostClassifier(estimator=e,n_estimators=o,random_state=25)
        ada.fit(X_train,y_train)
        y_pred_proba = ada.predict_proba(X_test)
        score.append([e,o,log_loss(y_test,y_pred_proba)])
score = pd.DataFrame(score,columns=['Estimator','Weak lerners','log_loss'])
score.sort_values('log_loss')
```

<h1 style = 'color : orange' >Adaboost Regresssor</h1>

```python
concrete = pd.read_csv('../Cases/Concrete_Strength/Concrete_Data.csv')

X = concrete.drop('Strength', axis =1)
y = concrete['Strength']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=25,test_size=0.3)

ada = AdaBoostRegressor(estimator=en, n_estimators=10,random_state=25)
ada.fit(X_train,y_train)

y_pred = ada.predict(X_test)
print(r2_score(y_test,y_pred))
```

```python
#for different estimators

concrete = pd.read_csv('../Cases/Concrete_Strength/Concrete_Data.csv')

X = concrete.drop('Strength', axis =1)
y = concrete['Strength']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=25,test_size=0.3)


weak_lnr = [10,15,25]
est_list = [lr,dtr,en]
score = []

for e in est_list:
    for o in weak_lnr:
        ada = AdaBoostRegressor(estimator=e,n_estimators=o,random_state=25)
        ada.fit(X_train,y_train)
        y_pred = ada.predict(X_test)
        score.append([e,o,r2_score(y_test,y_pred)])
score = pd.DataFrame(score,columns=['Estimator','Weak lerners','score'])
score.sort_values('score',ascending = False)
```
