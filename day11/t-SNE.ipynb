```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

<h1 style = 'color:Orange'>T SNE</h1>

```python
tsne = TSNE(random_state = 25).set_output(transform='pandas')


cancer = pd.read_csv('../Cases/Wisconsin/BreastCancer.csv',index_col=0)
y = cancer['Class']
X = cancer.drop('Class',axis = 1)

X_tsme = tsne.fit_transform(X)
tsne_data = X_tsme.copy()
tsne_data['Class'] = y
tsne_data
sns.scatterplot(data=tsne_data, x = 'tsne0',y = 'tsne1',hue='Class')
```
