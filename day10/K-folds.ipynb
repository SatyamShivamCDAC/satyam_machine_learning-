```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import KFold,cross_val_score,cross_validate,cross_val_predict,StratifiedKFold
from tqdm import tqdm
```

<h1 style = color:orange>Cross Validataion k folds</h1>

```python
lr = LinearRegression()
kfold = KFold(n_splits=5,shuffle=True,random_state=25)

pizza = pd.read_csv('../Datasets/pizza.csv')

X = pizza.drop('Sales',axis=1)
y = pizza['Sales']

cross_val_score(lr,X,y,cv = kfold).mean()
```

<h1>K-FOLD</h1>

```python
cancer = pd.read_csv('../Cases/Wisconsin/BreastCancer.csv',index_col=0)

lr = LogisticRegression()
kfolds = StratifiedKFold(n_splits=5,shuffle=True,random_state=25)

X = cancer.drop('Class',axis = 1)
y = cancer['Class']

print(cross_val_score(lr,X,y,cv=kfolds,scoring='neg_log_loss').mean())
print(cross_val_score(lr,X,y,cv=kfolds,scoring='accuracy').mean())
print(cross_val_score(lr,X,y,cv=kfolds,scoring='f1_micro').mean())
print(cross_val_score(lr,X,y,cv=kfolds,scoring='roc_auc').mean())
```

```python
#all solvers

cancer = pd.read_csv('../Cases/Wisconsin/BreastCancer.csv',index_col=0)

kfolds = StratifiedKFold(n_splits=5,shuffle=True,random_state=25)

X = cancer.drop('Class',axis = 1)
y = cancer['Class']

solvers = ['lbfgs','liblinear','newton-cg','sag','saga','newton-cholesky']
scores = []
Cs = np.linspace(0.001,4,20)


for s in tqdm(solvers):
    for c in Cs:
        lr = LogisticRegression(solver=s,C=c)
        scores.append([s,c,cross_val_score(lr,X,y,cv=kfolds,scoring='roc_auc').mean()])
scores = pd.DataFrame(scores,columns=['Solver','alpha','Score'])
scores.sort_values('Score',ascending = False)
```
